## 1. Normalize Modalities

```python
import json

# Load the modalities_modify.json file
with open('./modalities/modalities_modify.json', 'r', encoding='utf-8') as file:
    modalities_modify = json.load(file)

# Create a mapping from old modality names to new category titles
modality_mapping = {}
for category in modalities_modify['categories']:
    for modality in category['modalities']:
        modality_mapping[modality] = category['title']
```

## 2. Modify the modalities in Result

```python
# Load the updated_result_2.json file
with open('reformatted_result_2.json', 'r', encoding='utf-8') as file:
    updated_result = json.load(file)

# Function to update modalities in the data
def update_modalities(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'modalities':
                for modality in value:
                    if modality['title'] in modality_mapping:
                        modality['title'] = modality_mapping[modality['title']]
            else:
                update_modalities(value)
    elif isinstance(data, list):
        for item in data:
            update_modalities(item)

# Update the modalities in the updated_result data
update_modalities(updated_result)

# Save the updated result back to a file
with open('updated_modalities_result_2.json', 'w', encoding='utf-8') as file:
    json.dump(updated_result, file, ensure_ascii=False, indent=4)
```

```python
# Load the updated_result_2_modified.json file
with open('updated_modalities_result_2.json', 'r', encoding='utf-8') as file:
    updated_result = json.load(file)

# Set to store unique modalities
unique_modalities = set()

# Function to extract modalities from the data
def extract_modalities(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'modalities':
                for modality in value:
                    unique_modalities.add(modality['title'])
            else:
                extract_modalities(value)
    elif isinstance(data, list):
        for item in data:
            extract_modalities(item)

# Extract modalities from the updated_result data
extract_modalities(updated_result)

# Print all unique modalities
print("Unique modalities found in the updated_modalities_result_2.json:")
for modality in unique_modalities:
    print(modality)

# Check if all modalities have been replaced successfully
all_replaced = all(modality in modality_mapping.values() for modality in unique_modalities)
if all_replaced:
    print("All modalities have been replaced successfully.")
else:
    print("Some modalities have not been replaced successfully.")
```
