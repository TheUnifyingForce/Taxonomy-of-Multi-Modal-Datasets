## 1.2 Dataset: a schema.org type -> https://schema.org/Dataset

**url**: URL of the item.

**name**: The name of the item.

**alternateName**: An alias for the item.

**description**: A description of the item.

**variablesMeasured**: The variableMeasured property can indicate (repeated as necessary) the variables that are measured in some dataset, either described as text or as pairs of identifier and description using PropertyValue, or more explicitly as a StatisticalVariable.

**measurementTechnique**: A technique, method or technology used in an Observation, StatisticalVariable or Dataset (or DataDownload, DataCatalog), corresponding to the method used for measuring the corresponding variable(s) (for datasets, described using variableMeasured; for Observation, a StatisticalVariable). Often but not necessarily each variableMeasured will have an explicit representation as (or mapping to) an property such as those defined in Schema.org, or other RDF vocabularies and "knowledge graphs". In that case the subproperty of variableMeasured called measuredProperty is applicable.

The measurementTechnique property helps when extra clarification is needed about how a measuredProperty was measured. This is oriented towards scientific and scholarly dataset publication but may have broader applicability; it is not intended as a full representation of measurement, but can often serve as a high level summary for dataset discovery.

For example, if variableMeasured is: molecule concentration, measurementTechnique could be: "mass spectrometry" or "nmr spectroscopy" or "colorimetry" or "immunofluorescence". If the variableMeasured is "depression rating", the measurementTechnique could be "Zung Scale" or "HAM-D" or "Beck Depression Inventory".

If there are several variableMeasured properties recorded for some given data object, use a PropertyValue for each variableMeasured and attach the corresponding measurementTechnique. The value can also be from an enumeration, organized as a MeasurementMetholdEnumeration.

**sameAs**: URL of a reference Web page that unambiguously indicates the item's identity. E.g. the URL of the item's Wikipedia page, Wikidata entry, or official website.

**doi**: De-referencable identifiers; A standardized unique number given to many (but not all) articles, papers, & books, by some publishers, to identify a particular publication. 

**identifier**: The identifier property represents any kind of identifier for any kind of Thing, such as ISBNs, GTIN codes, UUIDs etc. Schema.org provides dedicated properties for representing many of these, either as textual strings or as URL (URI) links. See background notes for more details.

**author**: The author of this content or rating. Please note that author is special in that HTML 5 provides a special mechanism for indicating authorship via the rel tag. That is equivalent to this and may be used interchangeably.

**isAccessibleForFree**: A flag to signal that the item, event, or place is accessible for free. Supersedes free.

**dateModified**: The date on which the CreativeWork was most recently modified or when the item's entry was modified within a DataFeed.

**distribution**: A downloadable form of this dataset, at a specific location, in a specific format. This property can be repeated if different variations are available. There is no expectation that different downloadable distributions must contain exactly equivalent information (see also DCAT on this point). Different distributions might include or exclude different subsets of the entire dataset, for example.

**spatialCoverage**: The spatialCoverage of a CreativeWork indicates the place(s) which are the focus of the content. It is a subproperty of contentLocation intended primarily for more technical and detailed materials. For example with a Dataset, it indicates areas that the dataset describes: a dataset of New York weather would have spatialCoverage which was the place: the state of New York.

**provider**: The service provider, service operator, or service performer; the goods producer. Another party (a seller) may offer those services or goods on behalf of the provider. A provider may also serve as the seller. Supersedes carrier.

**funder**: A person or organization that supports (sponsors) something through some kind of financial contribution.

**temporalCoverage**: The temporalCoverage of a CreativeWork indicates the period that the content applies to, i.e. that it describes, either as a DateTime or as a textual string indicating a time period in ISO 8601 time interval format. In the case of a Dataset it will typically indicate the relevant time period in a precise notation (e.g. for a 2011 census dataset, the year 2011 would be written "2011/2012"). Other forms of content, e.g. ScholarlyArticle, Book, TVSeries or TVEpisode, may indicate their temporalCoverage in broader terms - textually or via well-known URL. Written works such as books may sometimes have precise temporal coverage too, e.g. a work set in 1939 - 1945 can be indicated in ISO 8601 interval format format via "1939/1945".

Open-ended date ranges can be written with ".." in place of the end date. For example, "2015-11/.." indicates a range beginning in November 2015 and with no specified final date. This is tentative and might be updated in future when ISO 8601 is officially updated. Supersedes datasetTimeInterval.

## 1.3 Chunk: df2

```python
# read file exclude the last unnamed column

def read_csv_in_chunks(file_path, chunk_size=50000):
    columns = pd.read_csv(file_path, nrows=0).columns.tolist()
    columns_to_use = columns[:-1]
    
    dtype={'funder': 'object',
       'measurementTechnique': 'object',
       'provider': 'object',
       'variablesMeasured': 'object',
       'isAccessibleForFree': 'object',
       'dateModified': 'object'}
    
    chunk_list = []
    for chunk in pd.read_csv(file_path, usecols=columns_to_use, dtype=dtype, chunksize=chunk_size, delimiter=','):
        chunk_list.append(chunk)
        #chunk.info(verbose=True)
    
    df = pd.concat(chunk_list, ignore_index=True)
    
    print(f"Concatenated DataFrame shape: {df.shape}") # final size
    
    # check if space or hiden character
    df.columns = df.columns.str.strip()
    print(f"Cleaned Columns: {df.columns.tolist()}")
    
    # check if duplicated
    if df.columns.duplicated().any():
        print("Warning: Duplicate column names found!")
        df.columns = pd.io.parsers.ParserBase({'names':df.columns})._maybe_dedup_names(df.columns)
        print(f"Deduplicated Columns: {df.columns.tolist()}") 
     
    return df

file_path = 'dataset_metadata_2020_10_16.csv'
df2 = read_csv_in_chunks(file_path)
```

```python
# find_duplicates_in_one_column

def find_duplicates_in_one_column(df, column_name):
    if column_name in df.columns:
        duplicated_series = df[column_name][df[column_name].duplicated()]
        return duplicated_series.value_counts()
    else:
        return f"Column '{column_name}' does not exist in the DataFrame."

column_name = 'distribution'
duplicates = find_duplicates_in_one_column(df1, column_name)

print(f"Column '{column_name}' has the following duplicate values:")
print(duplicates)
```

### 1.4.2 dask with concat: df4

```python
def read_csv_with_dask(file_path):
#     columns = pd.read_csv(file_path, nrows=0).columns.tolist()
#     columns_to_use = columns[:-1]
    
    dtype={'funder': 'object',
       'measurementTechnique': 'object',
       'provider': 'object',
       'variablesMeasured': 'object',
       'isAccessibleForFree': 'object',
       'dateModified': 'object'}
    
    #df = dd.read_csv(file_path, usecols=columns_to_use)
    chunks = []
    chunksize = 50000 
    
    i = 0
    with open(file_path, 'r', encoding='utf-8') as file:
        while True:
            try:
                chunk = pd.read_csv(file, nrows=chunksize, dtype=dtype)
        
                if len(chunk) < chunksize:
                    break 
                
                num_rows = (i+1) * len(chunk)
                print(f"Chunk {i+1}: Read {num_rows} rows")
                
                chunks.append(chunk)
                i += 1
                
            except pd.errors.ParserError as e:
                print(f"Error parsing CSV: {e}")
                
                continue
    
    df = dd.from_pandas(pd.concat(chunks, ignore_index=True), chunksize=chunksize)
    return df

file_path = 'dataset_metadata_2020_10_16.csv'
df4_dask = read_csv_with_dask(file_path)
```

```python
def read_csv(file_path):
    columns = pd.read_csv(file_path, nrows=0).columns.tolist()
    columns_to_use = columns[:-1]
    
    dtype = {
        'dateModified': 'object',
        'funder': 'object',
        'measurementTechnique': 'object',
        'provider': 'object',
        'variablesMeasured': 'object'
    }
    
    #df = dd.read_csv(file_path, usecols=columns_to_use)
    chunk_list = []
    chunk_size = 50000
    
    for chunk in pd.read_csv(file_path, usecols=columns_to_use, chunksize=chunk_size, delimiter=','):
        chunk_list.append(chunk)
    
    df = dd.from_pandas(pd.concat(chunk_list, ignore_index=True), chunksize=chunk_size)
    return df

file_path = 'dataset_metadata_2020_10_16.csv'
df5_dask = read_csv(file_path)
```

```python
def read_one_line_csv(file_path, start_row, end_row):
    df = pd.read_csv(file_path, skiprows=range(1, start_row), nrows=end_row-start_row, header=None)
    return df

file_path = 'dataset_metadata_2020_10_16.csv'
start_row = 350005
end_row = 350010
df4 = read_one_line_csv(file_path, start_row, end_row)
```

```python
from bs4 import BeautifulSoup
import re

# Convert data to DataFrame
df6 = pd.DataFrame(df2['description'].head(10), columns=["description"])

def is_html(text):
    """
    Determine whether the text is in HTML format
    """
    # If it contains HTML tags, it is considered to be HTML format.
    return bool(re.search(r'<[^>]+>', text))

def convert_to_text(html):
    """
    Convert HTML to plain text
    """
    soup = BeautifulSoup(html, 'html.parser')
    return soup.get_text(separator=' ', strip=True)

# Line-by-line output of text content
for idx, row in df6.iterrows():
    content = row['description']
    if is_html(content):
        content = convert_to_text(content)
    print(f"Row {idx}: {content}")
```
